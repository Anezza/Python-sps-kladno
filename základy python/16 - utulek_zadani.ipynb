# 1
Vytvořte třídu Utulek, který bude obsahovat tyto vlastnosti:

*   Název
*   Maximální kapacita
*   SeznamZvířat

A jeho metody budou:

*  getNazev
*  getMaxKapacita
*  getSeznamZvirat
*  setNazev
*  setMaxKapacita
*  pridatZvire
*  odeberZvire
*  ToString







```python
class Utulek:
    def __init__(self,nazev,maxkapacita,zvirata=None):
        self.nazev = nazev
        self.maxkapacita = maxkapacita
        if zvirata is None:
            self.zvirata = []
        else:
            self.zvirata = zvirata

    def pridej_zvire(self, zvire):
        if len(self.zvirata) < self.maxkapacita:
            if zvire in self.zvirata:
                print(f"{zvire} je již v útulku")
            else:
                self.zvirata.append(zvire)
                print(f"{zvire} bylo přidáno do útulku") 
        else:
            print("útulek je plný")
            

    def odeber_zvire(self, zvire):
        if zvire not in self.zvirata:
            self.zvirata.remove(zvire)
            print(f"{zvire} bylo zlikvidováno")
        else:
            print(f"{zvire} v útulku není")

    def __str__(self):
        return f"""
Útulek s názvem: {self.nazev}
MAximální kapacita: {self.maxkapacita}
Seznam zvířat v útulku: {self.zvirata}
"""
zv1 = "Jezevec"
Utulekbez = Utulek("Happy paws", 5)
Utuleks = Utulek("Animal Haven", 3, [zv1,"rybička"])
Utuleks.pridej_zvire("taxi")
Utulekbez.odeber_zvire("rybička")
```

# 2
Vytvořte třídu Zvire, která bude obsahovat tyto vlasnosti
* Jméno
* Hmotnost

a metody budou pro get a set a ToString a metodu zarvi()

```python
class Zvire:
  # Každé zvíře má své jméno a hmotnost
  def __init__(self, jmeno, hmotnost):
    if hmotnost < 0:
      raise ValueError("Hmotnost Zvířete nemůže být záporná")
    self.jmeno = jmeno
    self.hmotnost = hmotnost

  
  def get_jmeno(self):
    return self.__jmeno
  def set_jmeno(self, jmeno):
    self.__jmeno = jmeno

  def get_hmotnost(self):
    return self.__hmotnost
  
  def set_hmotnost(self, hmotnost):
    self.__hmotnost = hmotnost

  # Každé zvíře umí zařvat
  def zarvi(self):
    print("Huááá hlasitostí")

  def __str__(self):
    return "Volej na mě {} a vážím {}*tíhové_zrychlení".format(self.jmeno, self.hmotnost)

z1 = Zvire("Anička", 30)
print(z1)
z1.zarvi()
  
```

```python
#TODO Pes
class Pes(Zvire):
  # Při "výrobě" Psa nejprve vyrobíme zvíře a potom doplníme vlastnosti navíc
  def __init__(self, jmeno, hmotnost, hlasitost):
    Zvire.__init__(self, jmeno, hmotnost) # siper().__init__(jmeno,hmotnost)
    self.hlasitost = hlasitost

  # Pes řve jinka než obyčejné zvíře
  def zarvi(self):
    print(f"Haf haf hlasitostí {self.hlasitost}")

  # Zobrazení Psa je kombinace Zvířete a speciálních vlastností
  def __str__(self):
    return "Jsem pes hlasitý {} {}".format(self.hlasitost, Zvire.__str__(self))
```
