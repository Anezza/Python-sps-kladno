Vytvořte třídu Člověk, která bude mít atributy:


1.   jméno:str
2.   příjmení:str
3.   telefon:int
4.   email:str
5. adresa:str
6. pojišťovna:str

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše o této osobě




````python
class Clovek:
    def __init__(self,jmeno,prijmeni,telefon,email,adresa,pojistovna):
        self.jmeno = jmeno
        self.prijmeni = prijmeni
        self.telefon = telefon
        self.email = email
        self. adresa = adresa
        self.pojistovna = pojistovna
        #self.vek = 0
    def get_jmeno(self):
        return self.jmeno
    def set_jmeno(self, novejmeno):
        self.jmeno = novejmeno
        print("Jméno bylo změněno")

    def get_prijmeni(self):
        return self.prijmeni
    def set_prijmeni(self,noveprijmeni):
        self.prijmeni = noveprijmeni
        print("Příjmení bylo změněno")

    def get_telefon(self):
        return self.telefon
    def set_telefon(self, novytelefon):
        self.telefon = novytelefon
        print("Telefon byl změněn")

    def get_email(self):
        return self.email
    def set_email(self, novyemail):
        self.email = novyemail
        print("Email byl změněn")

    def get_adresa(self):
        return self.adresa
    def set_adresa(self, novaadresa):
        self.adresa = novaadresa
        print("Adresa byla změněna")

    def get_pojistovna(self):
        return self.pojistovna
    def set_pojistovna(self, novapojistovna):
        self.pojistovna = novapojistovna
        print("Pojišťovna byla změněna")
        
    def __str__(self):#ToString -> vypise vsechny informace o tride ve stringu
        return f"""
Jméno: {self.jmeno} 
Příjmení: {self.prijmeni}
Telefon: {self.telefon} 
Email: {self.email}
Adresa: {self.adresa} 
Pojišťovna: {self.pojistovna}
"""
    
clovek1 = Clovek("Aneta", "Veselá", 1546, "aneta@spskladno.cz", "Jahodová 12", "VZP")
clovek2 = Clovek("bfbhjt", "GRNJU", 46, "an@spskladno.cz", "Jahodová 1", "VZDR")
print(clovek1)
clovek1.set_prijmeni("Nesmysl")
print(clovek1)
print(clovek2.get_jmeno())
        
````
